### Part 1: One time pad review

Recall that a one time pad is an encryption technique that relies on the XOR operation for binary strings. Follow the exercises below to practice communicating a prime number using a one time pad. All of the code needed is in one cell below.

```python
#Used to convert a number, n, to a binary string
def toBinary(n):
    return bin(n)[2:]

#The input here, length, should be how long you want the output binary string to be.
def randBitString(length):
    k = ""
    for i in range(length):
        k += str(np.random.randint(0,2))
    return k

#Both input parameters are binary strings, m is the plaintext binary string and k is the key binary string.
def oneTimePad(m,k):
    return bin(int(m,2)^int(k,2))[2:].zfill(len(m))
```

### Part 2: One time pads with images

Now we will apply the one time pad to images. The code for this portion of the lab is located in the cell below. Do not forget to run that cell!

```python
#This function is used to import an image that is in the OneTimePad folder. The input parameter, name, should be a string and contain the extension of the file.
def importImage(name):
    return mpimg.imread(name)

#This function displays an image. The input parameter, img, should be the output of the importImage function.
def showImage(img):
    plt.xticks([]), plt.yticks([])
    plt.imshow(img)
    plt.show()
    
#This function performs an XOR operation on the pixels of an image. 
def imageXOR(img1,img2):
    return img1^img2

#This function returns a random collection of pixels of a specified dimension, w x h. 
def randomPad(w,h):
    return np.random.randint(256, size=(w,h,3), dtype='uint8')
```
